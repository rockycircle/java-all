```python
import pymysql 
import requests
import json
from bs4 import BeautifulSoup
url='http://www.kuaidi100.com/query?type=shentong&postid=3341815047094&id=1&valicode=&temp=0.6727097472263404' 
content=requests.get(url)
jd=json.loads(content.text)
i = 0
for ent in jd['data']:
    print(ent['context']);
    print(ent['time']);
    ida = i 
    content = ent['context'] 
    time = ent['time'] 
    i=i+1
    try:  
    # 获取数据库连接  
      conn = pymysql.connect(host='localhost', user='root', password='123456', db='logisticdb', port=3306, charset='utf8')  
      cur = conn.cursor()  # 获取一个游标  
      sql=" INSERT INTO logistic( d_id ,d_content , d_time ) VALUES (%s,%s,%s );"  
      print(sql) 
      cur.execute(sql,( ida , content , time  ))#加载sql语句  
      conn.commit()#提交事务  
      cur.close()  # 释放游标  
      conn.close()  # 释放资源  
    except Exception as e:  
        print("异常" + e)  

    
```
